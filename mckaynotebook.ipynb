```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint, solve_ivp, solve_bvp
from scipy.optimize import minimize
```

```python
beta=10
gamma=2
k=.6
#define ode measuring the spread of disease
def ode(t,y): 
    return np.array([beta*y[1]*y[0]*(1-y[0]/k),-beta*y[1]*y[0]*(1-y[0]/k)-gamma*y[1], gamma*y[1]])
#t vals
t0=0
tf=24
#initial conditions
y0=np.array([1,5/(3e6),0])
#solve ivp and get solution
solution=solve_ivp(ode,(t0,tf), y0, t_eval=np.linspace(t0,tf,150))
#plots and labels
plt.plot(solution.t, solution.y[0],label="In Ukraine")
plt.plot(solution.t, solution.y[1],label="Leaving")
plt.plot(solution.t, solution.y[2],'r',label="Left")
plt.xlabel('months')
plt.title('Ukraine Emigrant Population')
plt.ylabel('Proportion of Population')
plt.legend()
plt.show()
```
